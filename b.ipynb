```python
from scipy.stats import poisson

lead_time = 1 / 12 
target_beta = 0.95

lambda_1 = 1  
lambda_2 = 36   

# calculate
mu_1 = lambda_1 * lead_time
mu_2 = lambda_2 * lead_time

# hilffunktion
def calculate_stock_level(mu, target_beta):
    S = 0
    while poisson.cdf(S, mu) < target_beta:
        S += 1
    return S

# Compute base stock levels
S1 = calculate_stock_level(mu_1, target_beta)
S2 = calculate_stock_level(mu_2, target_beta)

# Compute achieved service levels
beta_1 = poisson.cdf(S1, mu_1)
beta_2 = poisson.cdf(S2, mu_2)

# Output results
print(f"S1={S1}, beta={beta_1:.4f}")
print(f"S2={S2}, beta={beta_2:.4f}")
```
