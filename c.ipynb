```python
from scipy.stats import poisson

lead_time = 1 / 12 

lambda_1 = 1
lambda_2 = 36

price_1 = 10000
price_2 = 1000

S1 = 1
S2 = 6

#calculate mu
mu_1 = lambda_1 * lead_time
mu_2 = lambda_2 * lead_time

# Individual betas
beta_1 = poisson.cdf(S1, mu_1)
beta_2 = poisson.cdf(S2, mu_2)

# Aggregate fill rate (value-weighted)
aggregated_beta = (lambda_1 * beta_1 +lambda_2 * beta_2) / (lambda_1 + lambda_2)

# Inventory investment
inventory_investment = S1 * price_1 + S2 * price_2

# Output
print(f"Beta 1 = {beta_1:.4f}")
print(f"Beta 2 = {beta_2:.4f}")
print(f"Aggregated beta = {aggregated_beta:.4f}")
print(f"Total Costs = {inventory_investment} EUR")
```
