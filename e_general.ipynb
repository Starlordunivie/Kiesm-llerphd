```python
from scipy.stats import poisson
import numpy as np

lead_time = 2 / 12
target_beta = 0.95

lambdas = np.array([15, 5, 2])
costs = np.array([1000, 3000, 2000])  
n = len(lambdas)

# calculate mu
mus = lambdas * lead_time


# Hilfsfunktion
def calculate_aggregated_beta(S):
    betas = poisson.cdf(S, mus)
    aggregated_beta = np.sum(lambdas * betas) / np.sum(lambdas)
    return aggregated_beta


# Greedy
S = np.zeros(n, dtype=int)
i = 0
current_beta = calculate_aggregated_beta(S)

while current_beta < target_beta:
    i += 1

    delta = np.zeros(n)

    # calcualte delta neu
    for j in range(n):
        S_plus = S.copy()
        S_plus[j] += 1

        beta_plus = calculate_aggregated_beta(S_plus)
        delta[j] = (beta_plus - current_beta) / costs[j]

    # wähle größtes delta
    j_star = np.argmax(delta)
    S[j_star] += 1

    # update beta
    current_beta = calculate_aggregated_beta(S)

    print(f"Iteration {i}: S={S}, Aggregated beta={current_beta}")

total_costs = np.sum(S * costs)
print(f"Total Costs = {total_costs} EUR")
```
