```python
from scipy.stats import poisson


lead_time = 1 / 12
target_beta = 0.95

lambda_1 = 1
lambda_2 = 36

c_1 = 10000
c_2 = 1000

# calculate mu
mu_1 = lambda_1 * lead_time
mu_2 = lambda_2 * lead_time


# hilfsfunktion
def calculate_aggregated_beta(S1, S2):
    beta_1 = poisson.cdf(S1, mu_1)
    beta_2 = poisson.cdf(S2, mu_2)

    aggregated_beta = (lambda_1 * beta_1 + lambda_2 * beta_2) / (lambda_1 + lambda_2)

    return aggregated_beta


# Greedy
S1 = 0 
S2 = 0
i = 0
current_beta = calculate_aggregated_beta(S1, S2)

while current_beta < target_beta:
    i += 1

    # calculate neu beta
    beta_1_plus = calculate_aggregated_beta(S1 + 1, S2)
    beta_2_plus = calculate_aggregated_beta(S1, S2 + 1)

    # calculate improvement
    delta_1 = (beta_1_plus - current_beta) / c_1
    delta_2 = (beta_2_plus - current_beta) / c_2

    # Choose Stock to increase by 1
    if delta_1 >= delta_2:
        S1 += 1
    else:
        S2 += 1

    # calculate neu beta 
    current_beta = calculate_aggregated_beta(S1, S2)

    print(f"Iteration {i}: S1={S1}, S2={S2}, Aggregated beta={current_beta}")


total_costs = S1 * c_1 + S2 * c_2
print(f"Total Costs = {total_costs} EUR")
```
