```python
from scipy.stats import poisson

lambda_1 = 2
lambda_2 = 3
lambda_3 = 5

S0 = 4
S1 = 1
S2 = 1
S3 = 2

# total drmand 
lambda_total = lambda_1 + lambda_2 + lambda_3


# in monaten
lead_time = 14 / 30

# erwartetet outstanding orders
mu = lambda_total * lead_time
print(mu)

k = 5

probability = poisson.pmf(k, mu)

print(probability)

```

```python
k = S0 - 1

probability = poisson.pmf(k, mu)

print(probability)
```

```python
k = S0

probability = poisson.cdf(k, mu)

print(probability)
```
